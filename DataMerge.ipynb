```python
import pandas as pd
```

```python
results = pd.read_csv("Data/results.csv")
plr = pd.read_csv("Data/playerattributes.csv")
```

```python
results = results[["Country", "Score", "Pos.", "Pts"]]
results
```

```python
plr = plr[["short_name","long_name", "nationality", "movement_acceleration", "movement_agility", "movement_balance", "power_jumping", "pace", "movement_reactions", "movement_sprint_speed", "power_stamina", "power_strength"]]
plr
```

```python
merged =  pd.merge(results, plr, left_on = 'Country', right_on = 'nationality')
merged
```

```python
# export the attributes merged
merged.to_csv("Data/merged_attributes.csv")
```

```python
grouped = merged[["Country", "Pos.", "Pts", "movement_acceleration", "movement_agility", "movement_balance", "power_jumping", "pace", "movement_reactions", "movement_sprint_speed", "power_stamina", "power_strength"]].groupby(["Country"]).mean()
```

```python
grouped
# export the mean of all the players' attributes grouped by their country
grouped.to_csv("Data/grouped_by_country.csv")
```

```python
# using sklearn

import statsmodels.api as sm
import numpy as np

# perform standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = grouped[["movement_acceleration", "movement_agility", "movement_balance", "power_jumping", "pace", "movement_reactions", "movement_sprint_speed", "power_stamina", "power_strength"]]
y = grouped["Pos."]
scaler.fit(X, y)
X_tr = scaler.transform(X)

# find the r value
X_tr = sm.add_constant(X_tr)
model = sm.OLS(y, X_tr).fit()
print_model = model.summary()
print(print_model)
```

```python
# using statsmodels

import statsmodels.api as sm
import numpy as np
from scipy import stats

# perform standardization
std_group = grouped.select_dtypes(include=[np.number]).dropna().apply(stats.zscore)
X = std_group[["movement_acceleration", "movement_agility", "movement_balance", "power_jumping", "pace", "movement_reactions", "movement_sprint_speed", "power_stamina", "power_strength"]]
y = std_group["Pos."]

# export standardized data to csv
results = pd.merge(y, X, on="Country")
results.to_csv("Data/standardized_merged_data.csv")

# find the r value
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print_model = model.summary()
print(print_model)
```
